```python
#import the necessary libraries
import pandas as pd
```

```python
#read the 5 rows
df=pd.DataFrame(data)
df.head(5)
```

```python
#draw a histogram and use more bins
df['A'].hist(bins=30)
```

```python
#make a scatter plot and modify the diagram for more visusal purpose
#s is size, figsize is length and width of the image
df.plot(kind='scatter',x='A',y='B',figsize=(30,5),color='red',s=50)
```

```python
df.plot(kind='hexbin',x='A',y='B',figsize=(7,5),gridsize=25)
```

```python
#box plot
df.plot(kind='box',colormap='coolwarm')
```
