#creating visulization using MatplotLib

```python
#Import Numpy for creating the Data Set of range 0 to 100 fro 3 Dimensions
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
x,y,z
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
plt.title("title")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
plt.xlabel("x-Axis")
plt.ylabel("y-Axis")
plt.plot(x,y,color='red')
ax2=fig.add_axes([0.2,0.5,0.2,0.2])
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.plot(x,y,color='red')
```

```python
fig=plt.figure()
axess1=fig.add_axes([0,0,1,1])
axess2=fig.add_axes([0.2,0.5,0.4,0.4])
```

```python
fig=plt.figure()
axess1=fig.add_axes([0,0,1,1])
axess1.plot(x,z)
plt.xlabel("x-axis")
plt.ylabel("z-axis")
axess2=fig.add_axes([0.2,0.5,0.4,0.4])
axess2.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("zoom")
plt.xlim([20,22])
plt.ylim([30,50])
```

```python
plt.subplots(ncols=2,nrows=1)
```

```python
fig,axes=plt.subplots(ncols=2,nrows=1,figsize=(12,2))
axes[0].plot(x,y,linestyle="--",color='blue',linewidth=3)
axes[1].plot(x,z,color='red',linewidth=3)
```
